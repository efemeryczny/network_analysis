```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Add nodes
G.add_node(1)
G.add_node(2)
G.add_node(3)

# Add undirected edges
G.add_edge(1, 2)
G.add_edge(2, 3)

# Add directed edges
G.add_edge(3, 1, direction='forward')
G.add_edge(1, 3, direction='backward')

# Set edge weights
G.edges[1, 2]['weight'] = 0.6
G.edges[2, 3]['weight'] = 0.8
G.edges[3, 1]['weight'] = 0.4
G.edges[1, 3]['weight'] = 0.2

# Draw the graph
pos = nx.spring_layout(G)  # Set the layout algorithm

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_size=200, node_color='lightblue')

# Draw undirected edges
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), arrows=False)

# Draw directed edges
directed_edges = [(u, v) for (u, v, d) in G.edges(data=True) if 'direction' in d]
nx.draw_networkx_edges(G, pos, edgelist=directed_edges, arrows=True)

# Draw edge labels
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# Set node labels
node_labels = {node: node for node in G.nodes()}
nx.draw_networkx_labels(G, pos, labels=node_labels)

# Show the graph
plt.axis('off')
plt.show()
```

```python
# Create a directed graph
directed_graph = nx.DiGraph()
directed_graph.add_node(1)
directed_graph.add_node(2)
directed_graph.add_edge(1, 2)

# Create an undirected graph
undirected_graph = nx.Graph()
undirected_graph.add_node(1)
undirected_graph.add_node(2)
undirected_graph.add_edge(1, 2)

# Create a weighted graph
weighted_graph = nx.Graph()
weighted_graph.add_node(1)
weighted_graph.add_node(2)
weighted_graph.add_edge(1, 2, weight=0.6)

# Create a graph with self-loop
self_loop_graph = nx.DiGraph()
self_loop_graph.add_node(1)
self_loop_graph.add_edge(1, 1)

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Plot directed graph
pos_directed = nx.spring_layout(directed_graph)
axs[0, 0].set_title("Directed Graph")
nx.draw_networkx(directed_graph, pos=pos_directed, with_labels=True, ax=axs[0, 0])

# Plot undirected graph
pos_undirected = nx.spring_layout(undirected_graph)
axs[0, 1].set_title("Undirected Graph")
nx.draw_networkx(undirected_graph, pos=pos_undirected, with_labels=True, ax=axs[0, 1])

# Plot weighted graph
pos_weighted = nx.spring_layout(weighted_graph)
axs[1, 0].set_title("Weighted Graph")
nx.draw_networkx(weighted_graph, pos=pos_weighted, with_labels=True, ax=axs[1, 0])

# Plot graph with self-loop
pos_self_loop = nx.spring_layout(self_loop_graph)
axs[1, 1].set_title("Graph with Self-loop")
nx.draw_networkx(self_loop_graph, pos=pos_self_loop, with_labels=True, ax=axs[1, 1])

# Adjust spacing between subplots
plt.tight_layout()

# Show the graphs
plt.show()
```

```python
# Create a directed graph
directed_graph = nx.DiGraph()
directed_graph.add_node('X')
directed_graph.add_node('Y')
directed_graph.add_edge('X', 'Y')

# Create an undirected graph
undirected_graph = nx.Graph()
undirected_graph.add_node('X')
undirected_graph.add_node('Y')
undirected_graph.add_edge('X', 'Y')

# Create a weighted graph
weighted_graph = nx.DiGraph()
weighted_graph.add_node('X')
weighted_graph.add_node('Y')
weighted_graph.add_node('Z')
weighted_graph.add_edge('X', 'Y', weight=0.2)
weighted_graph.add_edge('X', 'Z', weight=1)
weighted_graph.add_edge('Z', 'Y', weight=0.8)

# Create a graph with self-loop
self_loop_graph = nx.DiGraph()
self_loop_graph.add_node('X')
self_loop_graph.add_edge('X', 'X')

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Plot directed graph
pos_directed = nx.spring_layout(directed_graph)
axs[0, 0].set_title("Relacja skierowana")
nx.draw_networkx(directed_graph, pos=pos_directed, arrowsize=30, with_labels=True, ax=axs[0, 0], node_size=1000, font_size=12, font_weight='bold')

# Plot undirected graph
pos_undirected = nx.spring_layout(undirected_graph)
axs[0, 1].set_title("Relacja nieskierowana")
nx.draw_networkx(undirected_graph, pos=pos_undirected, with_labels=True, ax=axs[0, 1], node_size=1000, font_size=12, font_weight='bold')

# Plot weighted graph
pos_weighted = nx.circular_layout(weighted_graph)
axs[1, 0].set_title("Graf waony")
nx.draw_networkx(weighted_graph, pos=pos_weighted, with_labels=True, ax=axs[1, 0], node_size=1000, font_size=12, font_weight='bold')
edge_labels_weighted = nx.get_edge_attributes(weighted_graph, 'weight')
nx.draw_networkx_edge_labels(weighted_graph, pos=pos_weighted, edge_labels=edge_labels_weighted, ax=axs[1, 0])


# Plot graph with self-loop
pos_self_loop = nx.spring_layout(self_loop_graph)
axs[1, 1].set_title("Ptla")
nx.draw_networkx(self_loop_graph, pos=pos_self_loop, with_labels=True, ax=axs[1, 1], node_size=1000, font_size=12, font_weight='bold')

# Adjust spacing between subplots
plt.tight_layout()

# Show the graphs
plt.show()
```

```python
# Create a complete graph
complete_graph = nx.complete_graph(5)

# Add an additional node in the center
complete_graph.add_node('Center')
for node in complete_graph.nodes:
    if node != 'Center':
        complete_graph.add_edge('Center', node)

# Create a disconnected graph with at least 10 nodes
disconnected_graph = nx.Graph()
disconnected_graph.add_nodes_from(range(1, 12))
disconnected_graph.add_edges_from([(1, 2), (2, 3), (4, 5), (5, 6), (6, 7), (8, 9), (9, 10)])

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot complete graph
axs[0].set_title("Complete Graph")
pos_complete = nx.spring_layout(complete_graph)
nx.draw_networkx(complete_graph, pos=pos_complete, with_labels=True, ax=axs[0], node_color='lightblue', node_size=500, font_size=12, font_weight='bold', edge_color='gray')
nx.draw_networkx_edges(complete_graph, pos=pos_complete, ax=axs[0], edge_color='gray', width=1)

# Plot disconnected graph
axs[1].set_title("Disconnected Graph")
pos_disconnected = nx.spring_layout(disconnected_graph)
nx.draw_networkx(disconnected_graph, pos=pos_disconnected, with_labels=True, ax=axs[1], node_color='lightblue', node_size=500, font_size=12, font_weight='bold', edge_color='gray')
nx.draw_networkx_edges(disconnected_graph, pos=pos_disconnected, ax=axs[1], edge_color='gray', width=1)

# Adjust spacing
plt.tight_layout()

# Show the graphs
plt.show()
```
